### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type

urban_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]

suburban_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]

rural_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
```

```python
# 2. Get the total drivers for each city type
urban_driver_count = pyber_data_df.drop_duplicates(["city"]).groupby(["type"]).sum()["driver_count"]
    
suburban_driver_count = pyber_data_df.drop_duplicates(["city"]).groupby(["type"]).sum()["driver_count"]
    

rural_driver_count = pyber_data_df.drop_duplicates(["city"]).groupby(["type"]).sum()["driver_count"]
```

```python
#  3. Get the total amount of fares for each city type
urban_total_fare = pyber_data_df.groupby(["type"]).sum()["fare"]
suburban_total_fare = pyber_data_df.groupby(["type"]).sum()["fare"]
rural_total_fare = pyber_data_df.groupby(["type"]).sum()["fare"]
```

```python
#  4. Get the average fare per ride for each city type. 
urban_avg_fare = urban_total_fare / urban_ride_count
suburban_avg_fare = suburban_total_fare / suburban_ride_count
rural_avg_fare = rural_total_fare / rural_ride_count
```

```python
# 5. Get the average fare per driver for each city type. 
urban_fare_per_driver_avg = urban_total_fare / urban_driver_count
suburban_fare_per_driver_avg = suburban_total_fare / suburban_driver_count
rural_fare_per_driver_avg = rural_total_fare / rural_driver_count
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame({
                     
                    "Total Rides": urban_ride_count ,           
                    "Total Drivers": urban_driver_count,
                    "Total Fares": urban_total_fare,
                    "Average fare per Ride": urban_avg_fare, 
                    "Average fare per Driver": urban_fare_per_driver_avg })
pyber_summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average fare per Ride"] = pyber_summary_df["Average fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average fare per Driver"] = pyber_summary_df["Average fare per Driver"].map("${:,.2f}".format)
pyber_summary_df.head()
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 9. Using groupby() on the pyber_data_df to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_date_time_fare_df = pd.DataFrame(pyber_data_df.groupby(["type", "date"]).sum()[["fare"]])
total_date_time_fare_df
```

```python
# 10. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_date_time_fare_df = total_date_time_fare_df.reset_index()

```

```python
# 11. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city 
total_table =total_date_time_fare_df.pivot(index ='date', columns='type', values='fare')
total_table
```

```python
# 12. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2018-01-01':'2018-04-29'.

loc_df = total_table.loc['2019-01-01':'2019-04-29']
loc_df
```

```python
# 13. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
loc_df.index = pd.to_datetime(loc_df.index)
#loc_df.index = pd.to_datetime(loc_df)

```

```python
# 14. Check that the datatype for the index is datetime using df.info()
loc_df.info()
```

```python
# 15. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
loc_pivot= loc_df.resample("W").sum()
loc_pivot
```

```python
# 16. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

#Plotting sum_of_fare_by_weeks per city type uisng the Object-Oriented approach
ax = loc_pivot.plot(figsize=(20,10))

#Set axis 
ax.set_title("Total Fare by City Type", fontsize = "20")
ax.set_xlabel("Date", fontsize = "16")
ax.set_ylabel("Fare($USD)", fontsize = "16")
ax.grid(True)

ax.legend(["Urban","Suburban", "Rural", ], title="type", loc="center", fontsize="15", mode="Expanded")

plt.show()
```
